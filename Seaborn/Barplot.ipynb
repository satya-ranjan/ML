```python
import seaborn as sns
import numpy as np
```

```python
tips_df = sns.load_dataset("tips")
tips_df
```

```python
sns.barplot(x = tips_df.total_bill)
```

```python
sns.barplot(x = tips_df.day, y = tips_df.total_bill)
```

```python
sns.barplot(x = tips_df.day, y = tips_df.total_bill, hue = tips_df.sex) # to show the deta for male and female
```

```python
sns.barplot(x ="day", y = "total_bill", hue = "sex", data=tips_df)# if you dont want to write tips_df again and again you can use ---data---
```

```python
order = ["Sun" , "Thur" , "Fri" , "Sat"]# to change the order of X-axis

sns.barplot(x ="day", y = "total_bill", hue = "sex", data=tips_df,order= order)
```

```python
hue_order = ["Female" , "Male"]# IT will show first female and 2nd male  
order = ["Sun", "Thur" , "Fri" , "Sat"]

sns.barplot(x="day" , y = "total_bill" , hue="sex" , data= tips_df,order = order,hue_order=hue_order)
```

```python
sns.barplot(x="day" , y = "total_bill" , hue="sex" , data= tips_df, estimator = np.mean ) # to change the Y- axis as mean value
```

```python
sns.barplot(x="day" , y = "total_bill" , hue="sex" , data= tips_df, estimator = np.max ) # it will change the max value in y-axis which is 50 
```

```python
sns.barplot(x="day" , y = "total_bill" , hue="sex" , data= tips_df, estimator = np.max ,ci= 50) 
                                                                            # to change the black line
```

```python
sns.barplot(x="day" , y = "total_bill" , hue="sex" , data= tips_df, estimator = np.max ,ci= 50, n_boot=2)
```

```python
sns.barplot(x="day" , y = "total_bill" , hue="sex" , data= tips_df, orient= "h")
```

```python
sns.barplot(x="day" , y = "total_bill" , hue="sex" , data= tips_df, palette = "hot")
```

```python
sns.barplot(x="day" , y = "total_bill" , hue="sex" , data= tips_df, dodge= False) # It will show the graph in one
```
