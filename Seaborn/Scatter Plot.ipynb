# How to draw Seaborn Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_df = sns.load_dataset("titanic")
titanic_df
```

```python
sns.scatterplot(x = "age",y = "fare" , data= titanic_df)
```

```python
sns.scatterplot(x = "age",y = "fare" ,hue="sex", data= titanic_df) # To get more clear
```

```python
sns.scatterplot(x = "age",y = "fare" ,hue="sex", data= titanic_df,style="who") # To get more clear picture we have used ----style---
```

```python
sns.scatterplot(x = "age",y = "fare" ,hue="sex", data= titanic_df,style="who",size="who") #TO increase the size of who in the graph
```

```python
sns.scatterplot(x = "age",y = "fare" ,hue="sex", data= titanic_df,style="who",size="who",sizes= (100,400)) #TO increase the size of the points
```

```python
plt.figure(figsize=[16,9])
sns.scatterplot(x = "age",y = "fare" ,hue="sex", data= titanic_df,style="who",size="who",sizes= (100,400))
```

```python
plt.figure(figsize=[16,9])          # To see who are alive We have a coloum called "alive"
sns.scatterplot(x = "who",y = "fare" ,hue="alive", data= titanic_df,style="alive",size="who",sizes= (100,400))
```
