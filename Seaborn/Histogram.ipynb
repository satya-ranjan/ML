# Histogram

```python
import seaborn as sns
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
tips_df = sns.load_dataset("tips")
tips_df
```

```python
sns.distplot(tips_df["size"])
```

```python
sns.distplot(tips_df["total_bill"])
```

```python
sns.distplot(tips_df["total_bill"], bins=100)
```

```python
sns.distplot(tips_df["total_bill"], bins=100,)
```

```python
sns.distplot(tips_df["total_bill"], bins=100 ,rug=True)
```

```python
sns.distplot(tips_df["total_bill"],color="r")
```

```python
sns.distplot(tips_df["total_bill"],norm_hist = True,)
```

```python
sns.distplot(tips_df["total_bill"],axlabel="None",label="total bill",)

plt.title("Histogram of total Bill")
plt.legend()
plt.show()
```
