# Heatmap

Heatmap is the graphical representation of 2D data.Each data value represents i a matrix and it has a special colour

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Real world Dataset

```python
globalwarming_df = pd.read_csv('E:\\ML\\ML\\Seaborn\\Whoisresponsibleforglobalwarming.csv')
globalwarming_df
```

```python
globalwarming_df = globalwarming_df.drop(columns=["Country Code","Indicator Name","Indicator Code"],axis=1).set_index("Country Name")
globalwarming_df.head(6)
```

```python
plt.figure(figsize=[16,9])
sns.heatmap(globalwarming_df) # to see the heatmap 
```

```python
plt.figure(figsize=[16,9])
sns.heatmap(globalwarming_df,vmin=0,vmax=21) # It will set min and max value for this data
```

```python
plt.figure(figsize=[16,9])
sns.heatmap(globalwarming_df,vmin=0,vmax=21,cmap="binary") # It will change the colour
```

```python
plt.figure(figsize=[16,9])
sns.heatmap(globalwarming_df,vmin=0,vmax=21,annot= True)# IT will show the value in the graph
```

# --------------------------------------------------------------------------------------------------------

```python
plt.figure(figsize=[16,9])

annot_kws ={"fontsize": 10,
            "fontstyle":"italic",
            "color":"k",
            "rotation":0.6,
            "verticalalignment":"center",
            "backgroundcolor" : "w"}

sns.heatmap(globalwarming_df,vmin=0,vmax=21,annot= True,annot_kws= annot_kws)
```

```python
plt.figure(figsize=[16,9])
sns.heatmap(globalwarming_df,vmin=0,vmax=21,annot= True,linewidths=0.1)#It will change the line gap
```

```python
plt.figure(figsize=[16,9])
sns.heatmap(globalwarming_df,vmin=0,vmax=21,annot= True,linewidths=0.1,linecolor="k")#change the line colour
```

# Correlation Heatmap

```python
globalwarming_df.corr() # this is correlation with defult parameter
```

```python

sns.heatmap(globalwarming_df,annot =True,)
```

```python
plt.figure(figsize=[16,9])
ax= sns.heatmap(globalwarming_df.corr(),annot = True, linewidths= 0.2)
ax.tick_params(size = 10,color = "w",labelsize = 10)
plt.title("Heatmap of 'who is responcible for the golbal worming'")
plt.show()
```

# Second Dataset

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset =load_breast_cancer()
cancer_dataset
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_dataset["data"],cancer_dataset["target"]],
            columns= np.append(cancer_dataset["feature_names"],["target"]))

cancer_df
```

```python
plt.figure(figsize=(25,25))
ax = sns.heatmap(cancer_df.corr(),annot = True,linewidths =  3)
ax.tick_params(size = 10,color = "w",labelsize = 10)
plt.title("Heatmap of 'Breast Cancer Patients'",fontsize = 25)
plt.show()
```
