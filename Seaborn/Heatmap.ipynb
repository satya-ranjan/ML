# Heatmap

Heatmap is the graphical representation of 2D data.Each data value represents i a matrix and it has a special colour

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arr_2d = np.linspace(1,5,12).reshape(4,3)
arr_2d
```

```python
sns.heatmap(arr_2d)
```

# Real world Dataset

```python
globalwarming_df = pd.read_csv('E:\\ML\\ML\\Seaborn\\Whoisresponsibleforglobalwarming.csv')
globalwarming_df
```

```python
globalwarming_df = globalwarming_df.drop(columns=["Country Code","Indicator Name","Indicator Code"],axis=1).set_index("Country Name")
globalwarming_df.head(6)
```

```python
plt.figure(figsize=[16,9])
sns.heatmap(globalwarming_df) # to see the heatmap 
```

```python
plt.figure(figsize=[16,9])
sns.heatmap(globalwarming_df,vmin=0,vmax=21) # It will set min and max value for this data
```

```python
plt.figure(figsize=[16,9])
sns.heatmap(globalwarming_df,vmin=0,vmax=21,cmap="binary") # It will change the colour
```

```python
plt.figure(figsize=[16,9])
sns.heatmap(globalwarming_df,vmin=0,vmax=21,annot= True)# IT will show the value in the graph
```
