```python
Class = ["Python" , "R" , "Machine Learning" , " Artificial Intelligence" , "Data Science"]
Class1_Students = [45 , 15 , 35 , 25 , 30]

plt.pie(Class1_Students) #It will show the data
plt.show()
```

```python
Classes = ["Python" , "R" , "Machine Learning" , " Artificial Intelligence" , "Data Science"]
Class1_Students = [45 , 15 , 35 , 25 , 30]

plt.pie(Class1_Students,labels= Classes) #It will give the labels to the pie chart
plt.show()
```

```python
Classes = ["Python" , "R" , "Machine Learning" , " Artificial Intelligence" , "Data Science"]
Class1_Students = [45 , 15 , 35 , 25 , 30]
explode = [0.3,0,0.2,0,0]# give the explode parameter  0.2

plt.pie(Class1_Students,labels= Classes ,explode= explode,) #It will use here explode parameter
plt.show()
```

```python
Classes = ["Python" , "R" , "Machine Learning" , " Artificial Intelligence" , "Data Science"]
Class1_Students = [45 , 15 , 35 , 25 , 30]
explode = [0.3,0,0.2,0,0]
colors  = ["c","r","b","y","g"] 

plt.pie(Class1_Students,labels= Classes ,explode= explode,colors= colors) 
plt.show()
```

```python
Classes = ["Python" , "R" , "Machine Learning" , " Artificial Intelligence" , "Data Science"]
Class1_Students = [45 , 15 , 35 , 25 , 30]
explode = [0.3,0,0.2,0,0]
colors  = ["c","r","b","y","g"]

plt.pie(Class1_Students,labels= Classes ,explode= explode,colors= colors,autopct ="%0.2f%%") # It will give the percentahe in side pie chart
plt.show()
```

```python
Classes = ["Python" , "R" , "Machine Learning" , " Artificial Intelligence" , "Data Science"]
Class1_Students = [45 , 15 , 35 , 25 , 30]
explode = [0.3,0,0.2,0,0]
colors  = ["c","r","b","y","g"]

plt.pie(Class1_Students,labels= Classes ,explode= explode,colors= colors,radius = 1.4)# radius It will increase tge pi plot size
plt.show()
```

```python
Classes = ["Python" , "R" , "Machine Learning" , " Artificial Intelligence" , "Data Science"]
Class1_Students = [45 , 15 , 35 , 25 , 30]
explode = [0.3,0,0.2,0,0]
colors  = ["c","r","b","y","g"]
textprops = {"fontsize": 15}

plt.pie(Class1_Students,labels= Classes ,explode= explode,colors= colors,radius = 1.4,textprops= textprops)
plt.show()
```
