# How to show image using matplotlib imshow fuction ?

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread("pie_chart.png")
img
```

```python
type(img)
```

```python
img.shape
```

```python
img.ndim
```

```python
plt.imshow(img)
```

```python
plt.imshow(img)
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.axis("off") # It will stop showing axis
plt.imshow(img)
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.axis("off") # It will stop showing axis
plt.imshow(img)
plt.colorbar() # It will show color bar
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.axis("off") # It will stop showing axis
plt.imshow(img,cmap= "Accent")
plt.colorbar() # It will show color bar
plt.show()
```
