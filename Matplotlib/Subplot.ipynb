# Sub Plot

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(2,2,1)# to show first sub plot
plt.subplot(2,2,2)# To show second sub plot
plt.subplot(2,2,3)
plt.subplot(2,2,4)

plt.show()
```

```python
plt.subplot(2,2,1)

plt.pie([1]) #It will show pie plot in first sub plot

plt.subplot(2,2,2)

plt.pie([1,2,3]) #It will show pie plot in second sub plot

plt.subplot(2,2,3)

plt.pie([1,2,3,4])

plt.subplot(2,2,4)

plt.pie([1,2])


plt.show()
```
