# Ploting Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style 
```

```python
classes = ["Python", "R" , "AI" , "ML", "DS"]
class1_students = [30,10,20,25,10]
class2_students = [40,5,20,20,10]
class3_students = [30,5,30,15,15]
```

```python
plt.bar(classes,class1_students)# It will plot the bar chart 
```

```python
plt.barh(classes,class1_students) # TO print the bar in horigentaly
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge") # It will change the width size And change the align angle 
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color = "r") # It will chanage the bar color 
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color = "r", edgecolor = "y",linewidth = 5)
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color = "r", edgecolor = "y",linewidth = 5,linestyle ="--")
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,width=0.2,align="edge",color = "r", edgecolor = "y",
            linewidth = 5,linestyle ="--")
```

```python
plt.figure(figsize=(16,9))
plt.barh(classes,class1_students,width=0.2,align="edge",color = "r", edgecolor = "y",
            linewidth = 5,linestyle ="--") # with the width we can do horizental
```

```python
plt.figure(figsize=(16,9))
plt.barh(classes,class1_students,align="edge",color = "r", edgecolor = "y",
            linewidth = 5,linestyle ="--") # without width  we con do that
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,width=0.2,align="edge",color = "r", edgecolor = "y",
            linewidth = 5,linestyle ="--")

plt.title("Bar Chart of IAIP class",fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("Number of student",fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,width=0.2,color = "r",label ="class 1 student")
        
plt.bar(classes,class2_students,width=0.2,color = "b",label = "class 2 student")


plt.title("Bar Chart of IAIP class",fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("Number of student",fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(16,9))

class_index = np.arange(len(classes))
width = 0.2

plt.bar(class_index,class1_students,width,color = "r",label ="class 1 student")

plt.bar(class_index + width,class2_students,width,color = "g",label ="class 2 student") #After adding the width we can see the difference

plt.title("Bar Chart of IAIP class",fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("Number of student",fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(16,9))

class_index = np.arange(len(classes))
width = 0.2

plt.bar(class_index,class1_students,width,color = "r",label ="class 1 student")

plt.bar(class_index + width,class2_students,width,color = "g",label ="class 2 student") #After adding the width we can see the difference

plt.xticks(class_index+width, classes) # It will give the index value
plt.title("Bar Chart of IAIP class",fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("Number of student",fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(16,9))

class_index = np.arange(len(classes))
width = 0.2

plt.bar(class_index,class1_students,width,color = "r",label ="class 1 student")

plt.bar(class_index + width,class2_students,width,color = "g",label ="class 2 student") #After adding the width we can see the difference

plt.xticks(class_index+width, classes,rotation = "25") # It will give the index value and rotate the index
plt.title("Bar Chart of IAIP class",fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("Number of student",fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(16,9))

class_index = np.arange(len(classes))
width = 0.2

plt.bar(class_index,class1_students,width,color = "r",label ="class 1 student")

plt.bar(class_index + width,class2_students,width,color = "g",label ="class 2 student") #After adding the width we can see the difference

plt.bar(class_index + width +width,class3_students,width,color = "y",label ="class 3 student")

plt.xticks(class_index+width, classes,rotation = "25") # It will give the index value and rotate the index
plt.title("Bar Chart of IAIP class",fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("Number of student",fontsize = 15)
plt.show()
```
