# Ploting Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style 
```

```python
classes = ["Python", "R" , "AI" , "ML", "DS"]
class1_students = [30,10,20,25,10]
class2_students = [40,5,20,20,10]
class3_students = [30,5,30,15,15]
```

```python
plt.bar(classes,class1_students)# It will plot the bar chart 
```

```python
plt.barh(classes,class1_students) # TO print the bar in horigentaly
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge") # It will change the width size And change the align angle 
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color = "r") # It will chanage the bar color 
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color = "r", edgecolor = "y",linewidth = 5)
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color = "r", edgecolor = "y",linewidth = 5,linestyle ="--")
```
