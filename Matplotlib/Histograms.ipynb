# Ploting Histrograms

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age = np.random.randint(18,45, (100))
py_student_age = np.random.randint(15,40, (100))
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)
plt.title("ML Student age histograms ")
plt.xlabel("Student age categoty")
plt.ylabel("No. of student age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] # It will divided the graph range
plt.hist(ml_student_age,bins,rwidth=0.8) # rwidth help to give gap between the bar
plt.title("ML Student age histograms ")
plt.xlabel("Student age categoty")
plt.ylabel("No. of student age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype="bar",orientation="horizontal") # It will give side graph
plt.title("ML Student age histograms ")
plt.xlabel("Student age categoty")
plt.ylabel("No. of student age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype="bar",orientation="horizontal",color="r")
plt.title("ML Student age histograms ")
plt.xlabel("Student age categoty")
plt.ylabel("No. of student age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype="bar",orientation="horizontal",
            label="Ml student") # It will give side graph
plt.title("ML Student age histograms ")
plt.xlabel("Student age categoty")
plt.ylabel("No. of student age")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype="bar",orientation="vertical",
            label="Ml student") # It will give side graph
plt.hist(py_student_age,bins,rwidth=0.8,histtype="bar",orientation="vertical",
            label="py student")
plt.title("ML Student age histograms ")
plt.xlabel("Student age categoty")
plt.ylabel("No. of student age")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,histtype="bar",orientation="vertical",
            label=["Ml student","py student"],color=["r","y"])
# plt.hist(py_student_age,bins,rwidth=0.8,histtype="bar",orientation="vertical",
            # label="py student")
plt.title("ML Student age histograms ")
plt.xlabel("Student age categoty")
plt.ylabel("No. of student age")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]

plt.figure(figsize=[16,9]) # it will change the fig size
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,histtype="bar",orientation="vertical",
            label=["Ml student","py student"],color=["r","y"])

plt.title("ML Student age histograms ")
plt.xlabel("Student age categoty")
plt.ylabel("No. of student age")
plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use("ggplot") # It will change the background color

bins = [15,20,25,30,35,40,45]

plt.figure(figsize=(16,9))
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,histtype="bar",orientation="vertical",
            label=["Ml student","py student"],color=["r","y"])

plt.title("ML Student age histograms ")
plt.xlabel("Student age categoty")
plt.ylabel("No. of student age")
plt.legend()
plt.show()
```
