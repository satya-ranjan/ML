# What is pandas Pivot Table ?

The level in the pivot table will be stored in Multilndex object () on the index and coloum s of the result DataFrame

Syntax : df.pivot_table()

Parameters :

            values         Fill_value

            index           margins

            columns         dropna

            aggfunc

# Pandas pivot table

```python
import pandas as pd
```

```python
df = pd.read_csv("E:\ML\ML\Pandas\Movie_rating.csv")
df
```

```python
df.pivot_table(index= "Genre") # It will give the min value of the table
```

```python
df.pivot_table(index="Genre", columns="Year") # It will give the mean value year wise
```

```python
df.pivot_table(index="Genre", columns="Year", aggfunc= "count") # It will count how many value are there in each fuction
```

```python
df.pivot_table(index="Genre", columns="Year", aggfunc= "max") # It will give the maximun value
```

```python
df.pivot_table(index="Genre",columns="Year",aggfunc="sum") #It will give the sum value
```

```python
df.pivot_table(index="Genre", columns="Year",fill_value="100000") #It will fill the null value
```

```python
df.pivot_table(index="Genre", columns="Year",fill_value=0,margins=True)
```

```python
df.pivot_table(index="Genre", columns="Year",aggfunc="sum",fill_value="0",margins=True)
```
